##1. Analyse the sales trend of the Videogames:

```python
df2 = df.groupby(['Year_of_Release'])[["Global_Sales"]].sum().sort_values(by=['Global_Sales'],ascending=[False]).reset_index()
df2

#aggregate the sales of every year
d = {'Global_Sales': 'sum', 'Year_of_Release':'first'}
df_new = df2.groupby('Year_of_Release', as_index=False).aggregate(d).reindex(columns=df2.columns)

df_new
plt.figure(figsize=(20, 8))

#plot the values
plt.plot(df_new['Year_of_Release'], df_new['Global_Sales'])

plt.xlabel('Year of release')
plt.ylabel("Global sales")
plt.title('Sales trend over the years')
plt.xticks( range(1985,2017,1) )

plt.show()

#print(df_new.loc[df_new['Global_Sales'] ==354.25 ])
```

##2. Analyze the release trend of the Videogames:

```python
df3 = df.groupby(['Year_of_Release'])[['Name']].count().sort_values(by=['Name'], ascending=[False]).reset_index()
df3
#aggregate the number of releases of each company
d = {'Name': 'sum', 'Year_of_Release':'first'}  
df_new = df3.groupby('Year_of_Release', as_index=False).aggregate(d).reindex(columns=df3.columns)
df_new
df_new.sort_values(by=['Year_of_Release'])

plt.figure(figsize=(20, 8))

#plot the values
plt.plot(df_new['Year_of_Release'], df_new['Name'])   

plt.xlabel('Year of release')
plt.ylabel("Number of games produced")
plt.title('Releases trend over the years')
plt.xticks( range(1985,2017,1) )

plt.show()
```

## 4.Sales Analysis of the most popular platform in all regions

```python
#grouping the North America sales based on each platform
data1 = df.groupby("Platform")[["NA_Sales"]].sum().sort_values(by=['NA_Sales'],ascending=[False]).reset_index()
data1.rename(columns = {'Platform':'Platform_NA'}, inplace = True)

#Grouping the europe sales based on each platform
data2 = df.groupby("Platform")[["EU_Sales"]].sum().sort_values(by=['EU_Sales'],ascending=[False]).reset_index()
data2.rename(columns = {'Platform':'Platform_EU'}, inplace = True)

data3 = df.groupby("Platform")[["JP_Sales"]].sum().sort_values(by=['JP_Sales'],ascending=[False]).reset_index()
data3.rename(columns = {'Platform':'Platform_JP'}, inplace = True)

#Grouping the global sales based on each platform
data4 = df.groupby("Platform")[["Global_Sales"]].sum().sort_values(by=['Global_Sales'],ascending=[False]).reset_index()
data4.rename(columns = {'Platform':'Platform_Glob'}, inplace = True)

#Concatenating our datasets
data=pd.concat([data1,data2,data3,data4],axis=1)
data

fig, axes = plt.subplots(2, 2, figsize=(15, 12), sharey=True)
fig.suptitle('Platform Sales in each region')

sns.barplot(ax=axes[0][0], x='Platform_NA', y='NA_Sales', data=data)
axes[0][0].set_title('NA Sales graph', loc = 'left')


sns.barplot(ax=axes[0][1], x='Platform_EU', y='EU_Sales', data=data)
axes[0][1].set_title('EU Sales graph', loc = 'left')


sns.barplot(ax=axes[1][0], x='Platform_JP', y='JP_Sales', data=data)
axes[1][0].set_title('JP Sales graph', loc = 'left')

sns.barplot(ax=axes[1][1], x='Platform_Glob', y='Global_Sales', data=data)
axes[1][1].set_title('Global Sales graph', loc = 'left')
```

### Inference : It is seen that the consoles by Sony and Nintendo are the best-selling in all the regions. This graph also gives the idea that the popularity of video games in North America is higher than in other regions.

## 7. Downloaded rate of each genre in the period 2013 to 2015

```python
profits = df.loc[df['Year_of_Release'] > 2012]
profits1 = profits.loc[profits['Year_of_Release'] < 2016]
profits2 = profits1.loc[:,['Genre','Year_of_Release','Total_downloads']]

pf = profits.groupby(['Genre','Year_of_Release']).Total_downloads.sum()

#Finding the percentage of each genre that has been downloaded
def percentage(ax):
  all_heights = [[p.get_height() for p in bars] for bars in ax.containers]
  for bars in ax.containers:
      for i, p in enumerate(bars):
          total = sum(xgroup[i] for xgroup in all_heights)
          percentage = f'{(100 * p.get_height() / total) :.1f}'
          ax.annotate(percentage, (p.get_x() + p.get_width() / 2, p.get_height()), size=11, ha='center', va='bottom')

#plot values
ax3 = sns.countplot(x = 'Genre',hue = 'Year_of_Release',data = profits2)

percentage(ax3)

plt.tight_layout()
plt.xticks(rotation = 90,fontsize = 15)
plt.yticks(fontsize = 15)
plt.rcParams['figure.figsize'] = [20,10]
plt.title('Genre of number of games downloaded in the period 2013 to 2015', fontsize=20)
plt.xlabel('Genres', fontsize=15)
plt.ylabel('Total Downloads of each genre(in percentage)', fontsize=15)
plt.show()
plt.clf()
```

###Inference : The downloads of the action genre are very consistent in this period. Fighting and Puzzle genres are the next most popular genres in the period.

#8. Which genre has sold the most in a single year?

```python
year_sale_dx = df.groupby(by=['Year_of_Release', 'Genre'])['Global_Sales'].sum().reset_index()

year_sale = year_sale_dx.groupby(by=['Year_of_Release'])['Global_Sales'].transform(max) == year_sale_dx['Global_Sales']

year_sale_max = year_sale_dx[year_sale].reset_index(drop=True)

genre = year_sale_max['Genre']

plt.figure(figsize=(30, 18))
#plot values
g = sns.barplot(x='Year_of_Release', y='Global_Sales', data=year_sale_max)
index = 0

#Finding the maximum Sales produced by genre in each year
for value in year_sale_max['Global_Sales']:
    g.text(index, value + 1, str(genre[index] + '-' +str(round(value, 2))), color='#000', size=14, rotation= 90, ha="center")
    index = index + 1


plt.xticks(rotation=90,fontsize = 20)
plt.yticks(fontsize = 18)
plt.title('Popular genre of each year\n\n', fontsize = 30)
plt.xlabel('Year of Release', fontsize=20)
plt.ylabel('Global sales', fontsize=20)
plt.show()
```

### Inference : It is seen that Sports was the most popular genre in the year 2006 with the highest sales globally. No other genre in the later years could beat the sales record of sports. Also, we observe that Action is the most popular genre for most of the years.
